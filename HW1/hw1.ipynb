### 中心极限定理
设独立同分布的随机变量 $X_1,X_2,\dots,X_n$ 期望为 $\mu$，方差为 $\sigma^2>0$，则当 $n$ 充分大时：
$$\frac{\sum_{i=1}^n X_i - n\mu}{\sigma\sqrt{n}} \xrightarrow{d} N(0,1)$$
即样本均值近似服从标准正态分布。
